```python
plt.figure(figsize=(10,5))

df['Avg ratings'].value_counts().sort_index().plot(kind='bar')

plt.title('Distribution of Average Ratings')
plt.xlabel('Ratings')
plt.ylabel('Number of Restaurants')

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['City'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(df['Price'], bins=30)

plt.title('Distribution of Restaurant Prices')
plt.xlabel('Price')
plt.ylabel('Number of Restaurants')

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))

sns.heatmap(df[['Price', 'Avg ratings', 'Total ratings', 'Delivery time']].corr(),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')

plt.show()
```

```python
top_food = df['Food type'].value_counts().head(10)

plt.figure(figsize=(12,6))

top_food.plot(kind='bar')

plt.title('Top 10 Food Types')
plt.xlabel('Food Type')
plt.ylabel('Count')

plt.xticks(rotation=45)

plt.show()
```

# Swiggy Data Cleaning & Visualization Project

## Project Overview

This project focused on cleaning, analyzing, and visualizing a Swiggy restaurant dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. The objective was to understand restaurant trends, delivery patterns, ratings, pricing behavior, and food preferences across different cities.

## Data Cleaning Process

* Checked for missing values
* Verified duplicate records
* Inspected data types and dataset structure
* Investigated suspicious rating patterns and outliers
* Validated numerical columns such as price, ratings, and delivery time

## Key Findings

* No missing values or duplicate rows were found in the dataset.
* Kolkata contained the highest number of restaurants in the dataset, followed by Mumbai.
* Most restaurant prices were concentrated between ₹0–₹500, indicating affordability-focused customer behavior.
* Delivery times were mostly concentrated between 40–65 minutes.
* A significant spike was observed at the 2.9 rating level, suggesting possible placeholder or normalized rating values.
* Indian cuisine was the most common food category, followed by Chinese, North Indian, Fast Food, and South Indian cuisines.
* Correlation analysis showed weak relationships between delivery time, price, and ratings.

## Tools & Libraries Used

* Python
* Pandas
* Matplotlib
* Seaborn
* Google Colab

## Conclusion

The project provided practical experience in data cleaning, exploratory data analysis, and visualization techniques. It also helped in understanding how business insights can be extracted from raw datasets using Python-based analytical tools.
